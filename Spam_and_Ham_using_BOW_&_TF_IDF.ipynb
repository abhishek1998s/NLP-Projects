```python
import pandas as pd
message = pd.read_csv('spamhamdata.csv', sep='\t',names=["label","message"])
message.sample(10)
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

**Stemming and Lemmatization**

```python
ps=PorterStemmer()
lemmatizer=WordNetLemmatizer()
```

```python
corpus=[]
for i in range(0,len(message)):
  #removing non letter character
  review=re.sub('[^a-zA-Z]',' ',message['message'][i])
  #lowercase
  review=review.lower()
  #spliting the words
  review=review.split()
  #applying stemming and removing stopwords
  review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
  # Apply lemmatization
  review = [lemmatizer.lemmatize(word) for word in review]
  # Join words back into a single string
  review = ' '.join(review)
  # Append to corpus
  corpus.append(review)
```

```python
corpus
```

**Creating bag of words**

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=2500,ngram_range=(1,2))
#x is independent feature
x=cv.fit_transform(corpus).toarray()
```

```python
cv.vocabulary_
```

```python
y=pd.get_dummies(message['label'])
y=y.iloc[:,1].values
y
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_detect_model=MultinomialNB().fit(x_train,y_train)
```

```python
y_pred=spam_detect_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

**Creating TF-IDF and NGgram**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
 tfidf=TfidfVectorizer(max_features=2500,ngram_range=(1,2))
 x=tfidf.fit_transform(corpus).toarray()
```

```python
import numpy as np
np.set_printoptions(edgeitems=30,linewidth=100000,formatter=dict(float=lambda x: "%.3g" % x))
```

```python
x
```

```python
tfidf.vocabulary_
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_tfidf_model=MultinomialNB().fit(x_train,y_train)
```

```python
y_pred=spam_tfidf_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
